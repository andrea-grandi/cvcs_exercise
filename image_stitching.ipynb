```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

im_a = Image.open("question-data/gallery_0.jpg")
plt.imshow(im_a)
plt.show()

im_b = Image.open("question-data/gallery_0.jpg")
plt.imshow(im_b)
plt.show()
```

```python
pts_im_a = np.array([[193, 33], [180, 239], [317, 94],[310, 209]])
pts_im_b = np.array([[138, 50], [132, 194], [340, 59],[333, 199]])
```

```python
h, status = cv2.findHomography(pts_im_a, pts_im_b)
```

```python
im_out = cv2.warpPerspective(im_b, h, (im_a.shape[1], im_a.shape[0]))
```

```python
plt.imshow(im_out)
plt.show()
```
