```python
from io import BytesIO
import numpy as np
import cv2
import torch
from skimage import data

im = torch.from_numpy(data.coins()[160:230, 70:270])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(im)
```

```python
t_lower = 260
t_upper = 260*3
edges = cv2.Canny(im.numpy(), t_lower, t_upper)
plt.imshow(edges)
```

```python
circles = cv2.HoughCircles(edges, cv2.HOUGH_GRADIENT, dp=1.3, minDist=30, param1=260*3, param2=15, minRadius=17, maxRadius=24)
```

```python
out = im
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        center = (i[0], i[1])
        color = (255, 0, 0)
        out = cv2.circle(np.float32(out), center, i[2], color, 2)
```

```python
plt.imshow(out)
```
